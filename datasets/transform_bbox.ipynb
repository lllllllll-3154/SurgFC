```python
import pandas as pd
import numpy as np
import json
import os
import time
from tqdm import tqdm
import cv2
```

```python
mask_dir = "/home/lxj/dataset/Wales_anno_4604_origin/training/output"
mask_name = os.listdir(mask_dir)
```

```python
import cv2
def masks_to_bboxes(img):
    # 找出所有独特的mask值，假设背景的值为0

    mask = cv2.imread(img)
    mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
    unique_values = np.unique(mask)
    unique_values = unique_values[unique_values != 0]  # 移除背景值

    # 创建一个字典来存储每个物体的边界框
    bboxes = {}

    # 遍历每个独特的mask值
    for value in unique_values:
        # 获取当前物体的mask
        object_mask = mask == value

        # 找到当前物体mask中所有非零点的坐标
        pos = np.where(object_mask)
        
        # 如果当前物体mask为空，就跳过
        if pos[0].size == 0 or pos[1].size == 0:
            continue
        
        # 获取边界框的坐标
        x_min = np.min(pos[1])
        x_max = np.max(pos[1])
        y_min = np.min(pos[0])
        y_max = np.max(pos[0])
        
        # 生成边界框并添加到字典中
        bbox = (x_min, y_min, x_max, y_max)
        bboxes[value] = bbox
    
    return bboxes
```

```python
bbox_json = []

for name in mask_name:
    temp = {}
    file_path = os.path.join(mask_dir,name)
    mask = cv2.imread(file_path)
    bbox = masks_to_bboxes(mask)
    temp["file_path"] = file_path
    temp["file_name"] = name
    temp["bbox"] = bbox
    bbox_json.append(temp)
    
    
```
